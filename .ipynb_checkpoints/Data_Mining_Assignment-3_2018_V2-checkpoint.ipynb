```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn import manifold
from sklearn import metrics
from sklearn.ensemble import IsolationForest
from matplotlib.ticker import MaxNLocator
import seaborn as sns
```

**Q1: Apply the (simple) k-means algorithm to this data, in order to identify clusters of similar customers. You should follow the elbow method covered in, starting from 2 clusters. Show the "best" clusters you found and motivate your choice of $k$.**  
**Step 1:** Read the original data and show the first five records. According to the sample data, the attributes 'age', 'income' and 'childre' are numeric data, the attribute 'region' is nominal data, and the rest attributes are binary data.

```python
data = pd.read_csv('./Assign3_Customer_Profiling.csv',sep=',')
data[:5]
```

**Step 2:** Since most of attributes are binary data, I convert 'region' to binary data firstly, and then convert nominal data to binary as well, which is using 0 and 1 to represent original data.

```python
# Convert data 'region' to binary data, which will add more columns
data = data.join(pd.get_dummies(data.pop('region')))
# Get nominal attributes
nominalFields = data.columns[data.dtypes == 'object']
# Use LabelEncoder library to convert nominal to binary
lb_encoder = LabelEncoder()
for label in nominalFields:
    data[label] = lb_encoder.fit_transform(data[label])
data[:5]
```

**Step 3:** The attributes 'age', 'income' and 'children' are numeric data with different scales, which leads to unbalanced weights of attributes, so I use StandardScaler library to rescale the data to standardize the range of independent variables. The StandardScaler assumes data is normally distributed within each feature and will scale them such that the distribution is centred around 0, with a standard deviation of 1.

```python
ss = StandardScaler()
data = pd.DataFrame(ss.fit_transform(data), columns=data.columns)
data[:5].round(3)
```

**Step 4:** Use elbow method to find the best cluster number. The cluster number K I used starts from 2 to 20, and is evaluated by the sum of squard errors. The result is shown below, which can be seen that the best cluster number is 5.

```python
# Set the range of K
K = range(2,21,1)
# Train the cluster models with different K
KMs = [KMeans(n_clusters=k, init='k-means++', max_iter=1000, random_state=10).fit(data) for k in K]
# Collect the sum of squard errors of each cluster model
squared_errors = [ kms.inertia_ for kms in KMs ]

# Plot the result
ax = plt.figure(figsize=(6,3.5)).gca()
ax.plot(K, squared_errors,'o-')
ax.xaxis.set_major_locator(MaxNLocator(integer=True))
ax.grid(True)
ax.set_xlabel("The number of clusters")
ax.set_ylabel("The sum of squared errors")
```

**Step 5:** Since the data is a high dimensional data, it is hard to visulize the cluster result in a 2D figure. In order to show how pure the result is, I use the Multidimensional scaling (MDS) algorithm to rescale dimensions from 13 to 2. The result is shown below. From the figure ploted below, it looks pure.

```python
# Train the cluster model using 5 clusters and k-means++ initial method 
KMs = KMeans(n_clusters=5, init='k-means++', max_iter=1000, random_state=3).fit(data)
# Predict the labels of each record
labels = KMs.predict(data)
# Use MDS to rescale dimensions from 13 to 2
mds = manifold.MDS(2, max_iter=5000, eps=1e-9, n_init=1)
Y = mds.fit_transform(data)
# Store the new data into variable d with columns ['x','y','label']
dScaled = pd.DataFrame(Y, columns=['x','y'])
dScaled['label'] = labels

# Plot the result
fig = plt.figure(figsize=(5,3.5))
ax = fig.add_subplot(111)
scatter = ax.scatter(dScaled['x'], dScaled['y'], s=12, c=dScaled['label'])
ax.set_title('Number of components: ' + str(5))
plt.colorbar(scatter)
```

**Q2: In WEKA, the density-based clustering algorithm uses the simple k-means approach as a base learner. Explore whether you obtain "better" results when using this method, when compared t using k-means on its own.**  
**Step 1:** Use WEKA to cluster the scaled data in Q1 with cluster number 5, and save the result named SimpleKMeans_5.csv which is using SimpleKMeans algorithm, MakeDensityBasedClusterer_5.csv which is using MakeDensityBasedClusterer algorithm. For KMeans algorithm, I set the initial method as k-means++ instead of default Random method. Since WEKA uses 'ClusterX' to label each instance, I use LabelEncoder library to convert nominal labels to numeric labels which is using 0, 1, 2,... to represent each cluster.

```python
# Load the results of WEKA
simpleData_5 = pd.read_csv('./SimpleKMeans_5.csv',sep=',')
MakeDensityData_5 = pd.read_csv('./MakeDensityBasedClusterer_5.csv',sep=',')
# Convert label from 'ClusterX' to numeric label
simpleData_5['Cluster'] = lb_encoder.fit_transform(simpleData_5['Cluster'])
MakeDensityData_5['Cluster'] = lb_encoder.fit_transform(MakeDensityData_5['Cluster'])
```

**Step 2:** There are two methods to measure the quality of cluster models: Extrinsic method which is a supervised method, and Intrisic method which is a unsupervised method. Since the ground truth is unavailable, so I choose the second one, which evaluates the goodness of a clustering by considering how well the clusters are separated, and how compact the clusters are. I calculate Silhouette Coefficient for the model1 trained by SimpleKMeans and the model2 trained by MakeDensityBased. The Silhouette Coefficient results of them are shown below. From the result of Silhouette Coefficient, the model of MakeDensityBased is slightly better than SimpleMeans, with 0.112 comparing to 0.111. 

```python
# Caculate Silhouette Coefficient of SimpleKMeans
print "SimpleKMeans (5 clusters) Silhouette Coefficient: %0.3f" % \
      metrics.silhouette_score(simpleData_5.iloc[:,0:-1], simpleData_5['Cluster'])
# Caculate Silhouette Coefficient of MakeDensityBased
print "MakeDensityBased (5 clusters) Silhouette Coefficient: %0.3f" % \
      metrics.silhouette_score(MakeDensityData_5.iloc[:,0:-1], MakeDensityData_5['Cluster'])
```

**Step 3:** I use the data I scaled in Step 5 in Q1 and append the labeling results of two methods in order to visulize the cluster results of SimpleKMeans and MakeDensityBased with cluster 5.

```python
# Use MDS to rescale dimensions from 13 to 2
dSimple_5 = pd.DataFrame(Y, columns=['x','y'])
dSimple_5['label'] = simpleData_5['Cluster']

dMakeDensity_5 = pd.DataFrame(Y, columns=['x','y'])
dMakeDensity_5['label'] = MakeDensityData_5['Cluster']

# Plot the cluster result
fig = plt.figure(figsize=(10,3.5))
ax = fig.add_subplot(121)
scatter = ax.scatter(dSimple_5['x'], dSimple_5['y'], s=12, c=dSimple_5['label'])
ax.set_title('Number of clusters (SimpleKMeans): ' + str(5))
plt.colorbar(scatter)
ax = fig.add_subplot(122)
scatter = ax.scatter(dMakeDensity_5['x'], dMakeDensity_5['y'], s=12, c=dMakeDensity_5['label'])
ax.set_title('Number of clusters (MakeDensityBased): ' + str(5))
plt.colorbar(scatter)
```

**Step 4:** I use the DBSCAN to check whether I can get a better cluster model or not. Firstly, I iterate the parameter epsilon from 0.1 to 10 by increasing pace 0.1, and minimal number of points from 2 to 25 by increaing pace 1. Then I choose the best setting by calculating the Silhouette Coefficient. The best setting I got is: epsilon=2.9, minimal number of points=4, cluster number=4 and Silhouette Coefficient=0.196918. From the result, the Silhouette Coefficient of DBSCAN is better than both SimpleKMeans and MakeDensityBasedClusterer. Finaly, I also repeat the same process of Step 5 in Q1 to visulize the cluster results. According to the result of DBSCAN, clusters looks pure, and the purple nodes (label -1) are regarded as noise by DBSCAN.

```python
# Set the range of epsilon and minimal number of points
eps = np.arange(0.1,10,0.1)
ms = range(2,25,1)
# Initial the settings
efficient = 0
epsilon = 0
minSample = 0
n_clusters_ = 0

# Start iteration
for e in eps:
    for m in ms:
        db = DBSCAN(eps=e, min_samples=m, algorithm='ball_tree').fit(data)
        core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
        core_samples_mask[db.core_sample_indices_] = True
        # Calculate the number of clusters
        n_clusters = len(set(db.labels_)) - (1 if -1 in db.labels_ else 0)
        if n_clusters > 2:
            # Calculate the Silhouette Coefficient
            tmp = metrics.silhouette_score(data, db.labels_)
            if tmp > efficient:
                labels = db.labels_
                n_clusters_ = len(set(db.labels_)) - (1 if -1 in db.labels_ else 0)
                efficient = tmp
                epsilon = e
                minSample = m
            
print 'The best setting: eps: %f, min_samples: %d, Clusters: %d, Silhouette: %f' % \
      (epsilon, minSample, n_clusters_, efficient)

# Train the DBSCAN cluster model with the best setting
db = DBSCAN(eps=epsilon, min_samples=minSample, algorithm='ball_tree').fit(data)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
n_clusters_ = len(set(db.labels_)) - (1 if -1 in db.labels_ else 0)
print "DBSCAN (4 clusters) Silhouette Coefficient: %0.3f" % \
      metrics.silhouette_score(data, labels)

# Append the labeling results to rescaled dimensional dataset from 13 to 2 
# and visualize the cluster result
db = pd.DataFrame(Y, columns=['x','y'])
db['label'] = labels

fig = plt.figure(figsize=(5,3.5))
ax = fig.add_subplot(111)
scatter = ax.scatter(db['x'], db['y'], s=12, c=db['label'])
ax.set_title('Number of clusters (DBSCAN): ' + str(4))
plt.colorbar(scatter)
```

**Q3: Explain what a global outlier is, and apply any one of the techniques as discussed in class to identify a global outlier in the Customer data.  
(i) explain what outlier detection method you used, and  
(ii)discuss the properties (characteristics) of the global outlier you discouvered.**  
**1) ** **Global Outlier** is a data that deviates significantly from the all normal objects in the dataset as if it were generated by a different mechanism.  
**2) ** I use Isolation Forest method to detect the golbal outliers. Isolation Forest is a model-based method which is developing new models for high-dimensional outliers directely. The method explicitly isolates anomalies instead of constructing a profile of normal instances. Isolation Trees partition instances recursively until all instances are isolated, and then random partitioning produces noticeable shorter paths for anomalies. Hence, a forest with shorter path lengths of trees for some particular points are highly likely to be outliers.  
**3) ** I detect outliers in the dimensional scaled dataset I got in Q2. The result is shown below, where the yellow nodes are normal instances, the purple nodes are outliers. According to the figure shown below, the outliers are:  
* Outliers are all located at the edge of all instances.  
* Outliers are isolated because of long distances from normal instances, or high density cluster with few instances.

```python
# Isolation Forest
clf = IsolationForest(max_samples=450)
# Predict outliers of dScaled
clf.fit(dScaled.iloc[:,0:-1])
y_dScaled = clf.predict(dScaled.iloc[:,0:-1])

# Plot outliers
fig = plt.figure(figsize=(5,3.5))
ax = fig.add_subplot(111)
scatter = ax.scatter(dScaled['x'], dScaled['y'], s=12, c=y_dScaled)
ax.set_title('Outliers')
plt.colorbar(scatter)
```
